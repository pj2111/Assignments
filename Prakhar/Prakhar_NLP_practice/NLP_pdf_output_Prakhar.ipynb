```python
# try this method 

collection_dict={"Chapter":[],"Output":[],"Constraints":[],"Challenges":[]}
chapters = text.split("Chapter:")
for i, chap in enumerate(chapters):
    collection_dict['Chapter'].append(i)
    if "Output" in chap:
        idx_out_strt = chap.index("Output")
        if "Constraint" not in chap:
            if "Challange" not in chap:
                idx_out_end = len(chap)
            else:
                idx_out_end = chap.index("Challange")
        else:
            idx_out_end = chap.index("Constraint")
        # logging.Info(f"{chap[(idx_cons):(idx_chal)]}")
        collection_dict['Output'].append(chap[(idx_out_strt):(idx_out_end)])
    else:
        collection_dict['Output'].append("Notfound")
    if "Constraint" in chap:
        idx_cons = chap.index("Constraint")
        if "Challenge" not in chap:
            idx_cons_end = len(chap)
        else:
            idx_cons_end = chap.index("Challenge")
        # logging.Info(f"{chap[(idx_cons):(idx_chal)]}")
        # collection_dict['index']
        collection_dict['Constraints'].append(chap[(idx_cons):(idx_cons_end)])
    else:
        collection_dict['Constraints'].append("Notfound")
        
    if "Challenge" in chap:
        idx_chal = chap.index("Challenge")
        # logging.Info(f"{chap[(idx_cons):(idx_chal)]}")
        collection_dict['Challenges'].append(chap[(idx_chal):(len(chap))])
    else:
        collection_dict['Challenges'].append("Notfound")
        # print("===========")
# len(constraints)

# output > constraint > challenge > end 
```
