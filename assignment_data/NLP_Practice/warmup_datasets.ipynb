```python

def chunk_examples(examples):       # WAF to create chunks of questions column store in a new column # chunking = only 10 chars
    # print(examples['question'])
    chunks = []
    if isinstance(examples['question'], list):
        for sentence in examples["question"]:
            # print(sentence)
            # chunks = []
            chunks.append([sentence[i:i + 10] for i in range(0, len(sentence),10)])
    else:
        chunks += [examples['question'][i:i + 10] for i in range(0, len(examples['question']),10)]
    return {"chunked_questions": chunks}
```

```python
from random import randint
from transformers import pipeline

fillmask = pipeline("fill-mask", model="roberta-base")  # Creating a pipeline of fillmask model

mask_token = fillmask.tokenizer.mask_token

smaller_dataset = dataset.filter(lambda e, i: i<100, with_indices=True)  # lambda e
```

```python
# type(fillmask) --> transformers.pipelines.fill_mask.FillMaskPipeline
fillmask.tokenizer
```

```python
def augment_data(examples):
    outputs = []
    for sentence in examples["sentence1"]:  # looping through sentence1 in dataset
        words = sentence.split(' ')     # taking each word from sentence, ie chunking words
        K = randint(1, len(words)-1)    # selecting index of random word in sentence
        masked_sentence = " ".join(words[:K]  + [mask_token] + words[K+1:])  # inserting a mask after the random selected word
        predictions = fillmask(masked_sentence)     # applying fillmask
        augmented_sequences = [predictions[i]["sequence"] for i in range(3)]    # each sentence augmented in 3 ways
        outputs += [sentence] + augmented_sequences # added to outputs
    return {"data": outputs}
```

```python
augmented_dataset = smaller_dataset.map(augment_data,
                                        batched=True,
                                        remove_columns=dataset.column_names,
                                        batch_size=8)
augmented_dataset[:9]["data"]
```
