```python
# WAF to take file name, 
# create list of list with comma and 
# next line as sep and return shape of list

def shape_of_file(file_name):
    temp_list=[]
    with open(file_name, mode='r', encoding='utf-8') as file:
        f1 = file.read()
        rows = f1.split('\n')
        for row in rows:
            temp_list.append(row.split(","))
    return len(temp_list), len(temp_list[0])
```

```python
shape_of_file(wine_file)
```
